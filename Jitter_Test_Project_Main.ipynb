**Item-Specific Accuracy (Confusion Matrix)**

**Goal:** To prove that our model accurately distinguishes between different objects (e.g., person vs. backpack).

**What the Visual Shows:**
* **Diagonal:** Represents **Correct Detections (True Positives)**. For example, the high value where the "person" column meets the "person" row shows the model is highly accurate at identifying people.
* **Off-Diagonal:** Represents **Confusions (False Positives/Negatives)**. Any non-zero number off the diagonal shows a confusion. For instance, if there is a number in the "backpack" row under the "person" column, it means the model misidentified a person as a backpack some percentage of the time.

**Key Insight for Presentation:**
"Our robustness training resulted in low off-diagonal scores, meaning the model has a high level of **discrimination**â€”it rarely mistakes one object for another. We will zoom in on the specific items we are most interested in (e.g., vehicles) to show their near-perfect classification."
